# Descenso de Gradiente - Adam

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

Definimos la función de la que queremos encontrar el mínimo:

```python
def loss(phi0, phi1):
    height = np.exp(-0.5 * (phi1 * phi1) * 4.0)
    height = height * np.exp(-0.5 * (phi0 - 0.7) * (phi0 - 0.7) / 4.0)
    return 1.0 - height

# Calculamos los gradientes de esta función (por simplicidad, utilizamos diferencias finitas)
def get_loss_gradient(phi0, phi1):
    delta_phi = 0.00001
    gradient = np.zeros((2, 1))
    gradient[0] = (loss(phi0 + delta_phi / 2.0, phi1) - loss(phi0 - delta_phi / 2.0, phi1)) / delta_phi
    gradient[1] = (loss(phi0, phi1 + delta_phi / 2.0) - loss(phi0, phi1 - delta_phi / 2.0)) / delta_phi
    return gradient[:, 0]

# Calculamos la función de pérdida en un rango de valores de phi0 y phi1 para graficar
def get_loss_function_for_plot():
    grid_values = np.arange(-1.0, 1.0, 0.01)
    phi0mesh, phi1mesh = np.meshgrid(grid_values, grid_values)
    loss_function = np.zeros((grid_values.size, grid_values.size))
    for idphi0, phi0 in enumerate(grid_values):
        for idphi1, phi1 in enumerate(grid_values):
            loss_function[idphi0, idphi1] = loss(phi1, phi0)
    return loss_function, phi0mesh, phi1mesh
```

Definimos también un descenso de gradiente con paso fijo:

```python
def grad_descent(start_posn, n_steps, alpha):
    grad_path = np.zeros((2, n_steps+1));
    grad_path[:,0] = start_posn[:,0];
    for c_step in range(n_steps):
        this_grad = get_loss_gradient(grad_path[0,c_step], grad_path[1,c_step]);
        grad_path[:,c_step+1] = grad_path[:,c_step] - alpha * this_grad
    return grad_path;
```

Dado que la función cambia mucho más rápido en $\phi_1$ que en $\phi_0$, no hay un tamaño de paso ideal para elegir. Si el paso es pequeño, se necesitan muchas iteraciones hasta alcanzar el óptimo. Si incrementamos el paso, acabamos oscilando en torno al óptimo.

Esto motiva el uso de Adam, que tiene en consideración la magnitud de los gradientes en cada dirección.

Empecemos por normalizar las direcciones de los gradientes:
\begin{align}
\boldsymbol m_{t+1} \leftarrow & \frac{\partial \mathcal{L}[\boldsymbol\phi_t]}{\partial \boldsymbol\phi} \\
\boldsymbol v_{t+1} \leftarrow & \left(\frac{\partial \mathcal{L}[\phi_t]}{\partial \phi}\right)^2 \\
\boldsymbol\phi_{t+1} \leftarrow & \boldsymbol\phi_t
 - \alpha \cdot \frac{\boldsymbol m_{t+1}}{\sqrt{\boldsymbol v_{t+1}} + \epsilon}\end{align}

```python
def normalized_gradients(start_posn, n_steps, alpha,  epsilon=1e-20):
    grad_path = np.zeros((2, n_steps+1));
    grad_path[:,0] = start_posn[:,0];
    for c_step in range(n_steps):
        # Measure the gradient as in equation 6.13 (first line)
        m = get_loss_gradient(grad_path[0,c_step], grad_path[1,c_step]);
        #################################### COMPLETAR ####################################
        # Calcula el cuadrado del gradiente para v
        v = m**2
        # Actualiza el gradiente según la fórmula anterior
        grad_path[:,c_step+1] = grad_path[:,c_step]- alpha * (m /(np.sqrt(v) + epsilon))
        ################################## Fin COMPLETAR ##################################

    return grad_path;
```

Esto hace que nos movamos hacia el mínimo a una velocidad razonable, pero nunca llegamos a converger completamente: la solución simplemente rebota entre dos puntos.

Para lograr la convergencia, añadimos momento tanto a las estimaciones del gradiente como al gradiente cuadrado (energía).

Es decir:

\begin{align}
\boldsymbol m_{t+1} \leftarrow & \beta \cdot \boldsymbol m_{t} + (1-\beta)\cdot \frac{\partial \mathcal{L}[\boldsymbol\phi_t]}{\partial \boldsymbol\phi} \\
\boldsymbol v_{t+1} \leftarrow & \gamma \cdot\boldsymbol v_{t} + (1-\gamma)\cdot\left(\frac{\partial \mathcal{L}[\phi_t]}{\partial \phi}\right)^2 \\
\tilde{\boldsymbol m}_{t+1} \leftarrow & \frac{\boldsymbol m_{t+1}}{1-\beta^{t+1}} \\
\tilde{\boldsymbol v}_{t+1} \leftarrow & \frac{\boldsymbol v_{t+1}}{1-\gamma^{t+1}} \\
\boldsymbol\phi_{t+1} \leftarrow & \boldsymbol\phi_t
 - \alpha \cdot \frac{\tilde{\boldsymbol m}_{t+1}}{\sqrt{\tilde{\boldsymbol v}_{t+1}} + \epsilon}\end{align}

```python
def adam(start_posn, n_steps, alpha,  beta=0.9, gamma=0.99, epsilon=1e-20):
    grad_path = np.zeros((2, n_steps+1));
    grad_path[:,0] = start_posn[:,0];
    m = np.zeros_like(grad_path[:,0])
    v = np.zeros_like(grad_path[:,0])
    for c_step in range(n_steps):
        # Calculamos el gradiente
        grad = get_loss_gradient(grad_path[0,c_step], grad_path[1,c_step])

        #################################### COMPLETAR ####################################
        # Calcula el momento del gradiente (m)
        m = beta * m + (1-beta) * grad

        # Calcula el momento del su gradiente (v)
        v = gamma * v + (1-gamma) * grad**2

        # Actualiza m y v según las ecuaciones anteriores y el instante de tiempo (c_step)
        m_tilde = m / (1-beta**c_step+1)
        v_tilde = v /(1- gamma** c_step+1)


        # Actualiza, finalmente, el gradiente
        grad_path[:,c_step+1] = grad_path[:,c_step] - (alpha * m_tilde) / (np.sqrt(v) + epsilon)

    return grad_path;
```
